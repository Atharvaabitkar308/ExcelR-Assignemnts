```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt



# Before Standardization
plt.figure(figsize=(10, 6))
plt.title('Before Standardization')
plt.hist(df['Volume'], bins=30, color='skyblue', edgecolor='black', alpha=0.7, label='Volume')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.legend()
plt.grid(False)
plt.show()

# Standardization
scaler = StandardScaler()
df_standardized = df.copy()
df_standardized['Volume'] = scaler.fit_transform(df[['Volume']])

# After Standardization
plt.figure(figsize=(10, 6))
plt.title('After Standardization')
plt.hist(df_standardized['Volume'], bins=30, color='orange', edgecolor='black', alpha=0.7, label='Volume (Standardized)')
plt.xlabel('Standardized Volume')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
