```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data into a pandas DataFrame
df = pd.read_csv(r"C:\Users\athar\OneDrive\Desktop\python\Cardiotopographic.csv")
df
```

```python
df.isnull().sum()  # Display the count of missing values per column
```

```python
df.dtypes  # Display data types of each column
```

```python
# Descriptive statistics for each column
df.describe(include='all')  # Includes numerics and categoricals
```

```python
# Plot the boxplot for the LB column
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['LB'])
plt.title('Boxplot of Baseline Fetal Heart Rate (LB)')
plt.xlabel('Baseline Fetal Heart Rate (LB)')
plt.show()
```

```python

# Plot the boxplot for the MSTV column
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['MSTV'])
plt.title('Boxplot of Baseline Fetal Heart Rate (MSTV)')
plt.xlabel('Baseline Fetal Heart Rate (MSTV)')
plt.show()
```

```python
# Plot the boxplot for the MLTV column
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['MLTV'])
plt.title('Boxplot of Baseline Fetal Heart Rate (MLTV)')
plt.xlabel('Baseline Fetal Heart Rate (MLTV)')
plt.show()
```

```python
df.hist(figsize=(10, 6))
plt.tight_layout()
```

```python
df.boxplot(figsize=(10, 6))
plt.tight_layout()
```

```python
categorical_cols = [col for col in df.columns if df[col].dtype == 'object']
for col in categorical_cols:
  df[col].value_counts().plot(kind='bar')
  plt.title(col)
  plt.show()
```

```python
sns.pairplot(df)
plt.tight_layout()
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
