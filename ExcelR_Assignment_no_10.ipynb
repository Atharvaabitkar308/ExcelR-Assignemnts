```python
transaction=[]
for i in df.itertuples():
    l=set(list(i))
    transaction.append([i for i in l if (str(i)!='nan' and type(i)!=int)])
```

```python
transaction
```

```python
len(transaction)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from sklearn.preprocessing import OneHotEncoder
from mlxtend.frequent_patterns import apriori,association_rules,fpgrowth
```

```python
t=TransactionEncoder()
t_arr=t.fit_transform(transaction)
```

```python
data=pd.DataFrame(t_arr,columns=t.columns_)
data
```

```python
res=apriori(data, min_support=0.01, use_colnames=True)
res
```

```python
res=association_rules(res,metric='confidence', min_threshold=0.02)
res
```

```python
df['combined']=df.apply(lambda row:','.join(row.dropna()),axis=1)
```

```python
items=df['combined'].str.split(',')
```

```python
import pandas as pd
one_hot_encoded=pd.get_dummies(items.apply(pd.Series).stack()).groupby(level=0).sum()
```

```python
encoded_df=pd.concat([df,one_hot_encoded],axis=1)
```

```python
encoded_df
```

```python
item_counts=encoded_df.iloc[:,2:].sum()
item_counts
```

```python
sorted_items=item_counts.sort_values(ascending=False)
sorted_items
```

```python
N=12
plt.figure(figsize=(12,12))
sns.barplot(x=sorted_items[:N],
y=sorted_items.index[:N])
plt.xlabel('Frequency')
plt.ylabel('Food Item')
plt.title('Top {} Most popular Food Items'.format(N))
plt.show()
```

```python
co_occurrence_matrix=encoded_df.iloc[:,2:].T.dot(encoded_df.iloc[:,2:])
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(co_occurrence_matrix,cmap='viridis')
plt.xlabel('Food Item')
plt.ylabel('Food Item')
plt.title('Co-occurrence of Food Items')
plt.show()
```
